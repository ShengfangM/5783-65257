```python
#import needed library
import numpy as np
import matplotlib.pyplot as plt


#mount google drive on colab
from google.colab import drive
drive.mount('/content/drive')  

#both 'test.txt' and 'test.png' are stored in google drive
#the file path contain test files and save results
test_path = '/content/drive/MyDrive/Colab Notebooks/' #change to your path

#letters used to statistics/
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g',
           'h', 'i', 'j', 'k', 'l', 'm', 'n', 
           'o', 'p', 'q', 'r','s', 't', 
           'u', 'v', 'w', 'x','y', 'z'
           ]
#dictionary to store the histogram of letters
dict_letter=dict.fromkeys(letters, 0)    #letters as the key

#numpy array to store the frequency of each letter
arr_letter = np.zeros(len(letters),int)
```

```python
#open test file
f = open(test_path + 'test.txt','r') 
content = f.read()       #read the file
content_low = content.lower() #change all letter to lowercase
```

```python
#statistic the frequency of each letter
for (i,letter) in enumerate(letters):
  dict_letter[letter]= content_low.count(letter) 
  arr_letter[i] = dict_letter[letter]

#calculate the normalized frequency
arr_letter_norm = arr_letter/np.sum(arr_letter)

#print the histogram
print(dict_letter)
#print(arr_letter)
```

```python
#plot the histogram
#devide the plot area into two part
plt.subplot(2,1,1)
#plot the first bar with the statistic frequency
plt.bar(letters, arr_letter)
plt.show()
#plot the send bar with the normalized frequency
plt.subplot(2,1,2)

plt.bar(letters, arr_letter_norm)
plt.show()
```

```python
import cv2
img = cv2.imread(test_path + '/test.png') 

width, height, channels = img.shape
rgb = np.array([[255, 0, 0],[0, 255, 0],[0, 0, 255]])
#create 3D array to save the output data
outimg = np.zeros(img.shape, int)

plt.imshow(img[:,:,::-1])
```

```python
#question1
for i in range(width):
  for j in range(height):
    value = img[i,j,:]
# calculating Euclidean distance to the three points
# using linalg.norm()
    ECdists = np.linalg.norm(value -rgb,axis = 1)

    outimg[i,j,:] = rgb[(np.argmin(ECdists)),:]
    #print(dist_red,dist_green,dist_blue)
cv2.imwrite(test_path + 'output1.png', outimg) 

plt.imshow(outimg[:,:,::-1])
```

```python
#question 2
# find center point in the picture
centerpoint = np.array([width//2, height//2])
[ulx, uly] = centerpoint - 25

img2 = img
img2[ulx:ulx+50, uly:uly+50,:] = [0,0,0]

cv2.imwrite('output2.png', img2)

plt.imshow(img2[:,:,::-1])
```
